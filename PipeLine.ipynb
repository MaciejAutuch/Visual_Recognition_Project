```python
# function to trim the photos, it takes the label data frame, orginal image pathway, new image pathway and information about df rows and index
# add an error handling for pictures so that there is no problem with the picture size 

def trim_function(df, op_path, sv_path, row, nr):
  path = op_path + row['img_name']
  img = cv2.imread(path) # opening the photo and changing to array to make changes
  x = int(row['x_face_min'])
  y = int(row['y_face_min'])
  w = int(row['face_width'])
  h = int(row['face_height'])
  new_img = img[y:y+h, x:x+w] # using previously defined coordinates to trim the photo
  new_img = cv2.resize(new_img, (32, 32), interpolation = cv2.INTER_AREA)
  return cv2.imwrite(str(sv_path + 'train_' + str(nr) + '.jpg'), img) # returning a trimmed face picture
```

```python
#Processing the test photos

for index, row in test_label.iterrows():
  try:
    trim_function(test_label, '/content/test_images/images/', '/content/test_images_processed/', row, index)
  except:
    continue
```

```python
#Processing the train photos

for index, row in train_label.iterrows():
  try:
    trim_function(train_label, '/content/train_images/images/', '/content/train_images_processed/', row, index)
  except:
    continue
```

```python
# create mask label in test dataset

condition = [(test_label['occ_type'] == 1) |
             ((test_label['occ_type'] == 2) & (test_label['occ_degree'] == 3))]

value = [1]
test_label['face_mask'] = np.select(condition, value)
test_label.head(5)
```

```python
# create mask label in train dataset

condition = [(train_label['occ_type'] == 1) |
             ((train_label['occ_type'] == 2) & (train_label['occ_degree'] == 3))]

value = [1]
train_label['face_mask'] = np.select(condition, value)
train_label.head(5)
```
